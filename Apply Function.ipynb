# In this tutorial , we will learn about power of apply function. We will use big mart sales data.

```python
import pandas as pd
```

```python
df = pd.read_csv("bigmart_data.csv")
```

```python
df.shape
```

```python
# to get all rows
df.apply(lambda x:x)
```

```python
# to get row1
df.apply(lambda x:x[0])
```

```python
# to get column 1
df.apply(lambda x:x[0],axis=1)
```

```python
df["Item_MRP"].head()
```

```python
# creating function to clip all MRP greater than 200
def clip_mrp(mrp):
    if mrp > 200:
        mrp = 200
    return mrp


df["Item_MRP"].apply(lambda x:clip_mrp(x))[:5]        
```

```python
df["Outlet_Location_Type"].unique()
```

```python
def label_enc(loc_type):
    if loc_type == 'Tier 1':
        label = 1
    elif loc_type == 'Tier 2':
        label =2
    else:
        label =3
    return label
        
```

```python
df["Outlet_Location_Type"] = df["Outlet_Location_Type"].apply(label_enc)
```
