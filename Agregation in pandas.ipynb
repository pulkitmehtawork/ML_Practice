# We will learn about following 3 agreegate functions in this tutorial .
- groupby
- crosstab
- pivotable

```python
import pandas as pd
```

```python
df = pd.read_csv("bigmart_data.csv")
```

```python
df.head()
```

```python
# mean by item type

df.groupby("Item_Type").Item_MRP.mean()
```

```python
# cross tab
pd.crosstab(df["Outlet_Size"],df["Outlet_Location_Type"],margins=True)
```

```python
# Pivot table to get mean of sales for establishment year


pd.pivot_table(df,index = "Outlet_Establishment_Year",values = "Item_Outlet_Sales")
```
