# Tutorial to learn basics of matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Plot between Weight & Height
height = [160,170,180,190]
weight = [60,65,70,75]
```

```python
# Setting title & x, y axis labels

plt.title("Relationship between Height & Weight")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.plot(height,weight)


```

```python
# Labels & ticks 
plt.figure(figsize=(15,5))

calories = [65,70,75,80]
plt.plot(calories,'go')
plt.plot(weight,'y--')
plt.legend(labels = ["Calories Burnt","Weight"],loc = "lower right")
#plt.xticks(ticks = [0,1,2,3], labels = ["p1","p2","p3","p4"])
plt.xticks([0,1,2,3], ['p1', 'p2', 'p3', 'p4']);
```

```python
# Figure , subplot, sharey

fig, ax  = plt.subplots(nrows=1,ncols=2,figsize = (15,5),sharex = True,sharey = True)
ax[0].plot(calories,"go")
ax[1].plot(weight,"y--")
ax[0].set_title("Calories Burnt")
ax[1].set_title("Weight")
ax[0].set_xticks([0,1,2,3])
ax[1].set_xticks([0,1,2,3])
ax[0].set_xticklabels(['p1','p2','p3','p4'])
ax[1].set_xticklabels(['p1','p2','p3','p4'])

```
